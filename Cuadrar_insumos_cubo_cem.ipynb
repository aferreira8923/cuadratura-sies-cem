```python

def df_from_redshift(sql, engine):
    with engine.connect().execution_options(autocommit=True) as conn:
        query = conn.execute(text(sql))
        df = pd.DataFrame(query.fetchall())
    return df
```

## cem-matriculados

```python
años = ['2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021','2022','2023','2024']

df1 = pd.DataFrame(columns=['clasificacion_institucion_nivel_3','total_insumo','periodo','total_cubo','diferencias'])

texto="""
        ###########################
                Matriculados por tipo institución
        ###########################
        """
print(texto)

for año in años:
        print('###########################')
        print('######   AÑO: {} ######'.format(año))
        print('###########################')

        sql_insumo = '''SELECT tipo_inst_3, count(distinct mrun) as total_insumo 
                        FROM cem.cem_mat_p_{}
                        GROUP BY tipo_inst_3
                        ORDER BY tipo_inst_3;'''.format(año)

        df_insumo = df_from_redshift(sql_insumo, engine_insumo)
        
        df_insumo['periodo'] = [año] * len(df_insumo)
        df_insumo['periodo'] = df_insumo['periodo'].astype(int)
        df_insumo = df_insumo.rename(columns={'tipo_inst_3':'clasificacion_institucion_nivel_3'})

        sql_cubo = '''SELECT a.periodo,b.clasificacion_institucion_nivel_3,count(distinct mrun) as total_cubo
              FROM estudiantes.fact_matriculados as a
              LEFT JOIN transversal.dim_institucion as b 
              ON a.codigo_institucion = b.codigo_institucion AND a.periodo = b.periodo
              WHERE a.periodo = {}
              GROUP BY a.periodo,b.clasificacion_institucion_nivel_3;'''.format(año)
        
        df_cubo = df_from_redshift(sql_cubo, engine_cubo)

        df = df_insumo.merge(df_cubo,on=['periodo','clasificacion_institucion_nivel_3'])
        df['diferencias'] = df['total_insumo'] - df['total_cubo']

        df1 = pd.concat([df1,df])

df1.loc[(df1.loc[:,'diferencias']!= 0.)].to_csv('cem_mat_tipo_inst.csv',index=False)
```
